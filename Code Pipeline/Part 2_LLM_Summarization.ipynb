```python
import json
from pydantic import BaseModel
```

```python
class WordCountJSON(BaseModel):
    company_name: str
    word_count: int
    first_words: str
    last_words: str

test_lengths = [700, 750, 800, 1000]

for length in test_lengths:
    print(f"\n Testing with {length} tokens of input ")

    words = s.split()[:length]
    test_input = " ".join(words)

    first_10_words= " ".join(words[:10]) if len(words) >= 10 else " ".join(words)
    last_10_words = " ".join(words[-10:]) if len(words) >=10 else " ".join(words)

    try:
        response = ollama.chat(
            model="gemma3n:e2b",
            format = WordCountJSON.model_json_schema(),
            messages=[
                {
                    "role":"user",
                    "content": f"""Analyze this text and provide:
                    1. The company name mentioned in the text
                    2. The exact number of words
                    3. The first 10 words of the text
                    4. The last 10 words of the text

                    ---
                    {test_input}
                    ---"""
                } ],
        )

        result = json.loads(response['message']['content'])

        # Get word count and calculate accuracy
        actual_word_count = len(words)
        reported_count = result.get('word_count', 0)
        count_accuracy =(reported_count/actual_word_count * 100) if actual_word_count > 0 else 0

        saw_beginning = any(word in result.get('first_words', '').lower() for word in first_10_words.lower().split())
        saw_end = any(word in result.get('last_words', '').lower() for word in last_10_words.lower().split()[-3:])

        print(f"\n {length} words test:")
        print(f" Company : {result.get('company_name', 'Not found')}")
        print(f" Reported count : {reported_count} (Accuracy: {count_accuracy:.1f}%)")
        print(f" Saw beginning: {saw_beginning}")
        print(f" Saw end: {saw_end}")

        if count_accuracy < 80 or not saw_end:
            print(f" Model may be truncating at ~{length} words")

    except json.JSONDecodeError as e:
        print(f"\n JSON parse error at{length} words: {e}")
        print(f" Raw response: {response['message']['content'][:200]}..")
    except KeyError as e:
        print(f" \n Response structure error at {length} words: {e}")
        print(f" Response keys: {response.keys() if response in locals() else 'No response'}")
    except Exception as e:
        print(f"\n Failed at {length} words: {type(e).__name__}: {e}")
        break

print("\n" + "="*50)
print("Context Window Test Complete")
```

```python
from typing import List
! pip install llama-index

# Defining JSON schemas for junior and senior analyst
class ChunkAnalysis(BaseModel):
    company_name: str
    key_points: List[str]
    chunk_number: int
    has_material_info: bool

class FinalSummary(BaseModel):
    company_name: str
    business_description: str
    material_points: List[str]
    investment_industry: List[str]
    investment_exposure: List[str]




```

```python
from llama_index.core.node_parser import SentenceSplitter
from llama_index.core import Document

def analyze_chunks(document: str, ticker: str, model: str ="gemma3n:e4b") -> dict:
    """Analyze document in chunks and synthesize results."""

    # Create LlamaIndex Document
    doc = Document(text=document)

    chunker = SentenceSplitter(
    chunk_size= 700, # Note: LlamaIndex often defaults to tokens, not words
    chunk_overlap=50 # Overlap is also in tokens
    )
    # Get chunks (returns Node objects)
    chunk_nodes = chunker.get_nodes_from_documents([doc])

    # chunks = chunker(document,chunk_size=600, overlap=50)
    print(f"Document split into {len(chunk_nodes)} chunks")

    #Analyzing each chunk
    all_key_points =[]
    chunk_analyses = []

    for i, chunk in enumerate(chunk_nodes, 1):
        try:
            # Access the text using .text attribute
            chunk_text = chunk.text
            response = ollama.chat(
                model=model,
                format = ChunkAnalysis.model_json_schema(),
                options ={
                    "num_predict": 2500,
                    "temperature": 0.0,
                    "repeat_penalty": 1.2
                },
                messages=[
                    {
                        "role": "system",
                        "content": "You are a junior equity analyst extracting material information from company documents"
                    },
                    {
                        "role": "user",
                        "content": f"""Analyze this section of a company document and extract only the most ONLY the most material, stock-relevant information.
                        Your goal is:
                        - extract 1-2 bullet points as a summary of material information
                        - focus only on: Strategy, competitive position, revenue model, legal/regulatory matters, M&A activity, market advantages, and executive/governance changes.
                        - Prioritize recent event and growth drivers (2024-2025)
                        - Set has_material_info to false if no stock-relevant information is found.

                        Section {i} of {len(chunk_nodes)}:

                        ---
                        {chunk_text}
                        ---"""
                    }
                ]
            )
            # Parsing response
            chunk_result = json.loads(response['message']['content'])
            chunk_analyses.append(chunk_result)

            # Extracting key points if material info is found
            if chunk_result.get('has_material_info', False):
                new_info =chunk_result.get('key_points', [])
                all_key_points.extend(new_info)
                print(f" Chunk{i}: Found material points {new_info}")
            else:
                print(f" Chunk{i}: No material information")

        except Exception as e:
            print(f" Chunk {i}: Error - {e}")
            continue

    # Combining key points into a final summary

    # Check if we have any material information
    if not all_key_points:
        return {
            "status": "error",
            "message": "No material information found in document",
            "ticker": ticker
        }

    combined_points = "\n".join([f"• {point}" for point in all_key_points])
    print((f"\n Total key points collected:{len(all_key_points)}"))

    try:
        print("entered senior analyst try block")
        final_response= ollama.chat(
            model=model,
            format = FinalSummary.model_json_schema(),
            options ={
                "num_predict": 2500,
                "temperature": 0.0,
                "repeat_penalty": 1.2
            },
        messages=[
            {
                "role": "system",
                "content": "You are a senior equity analyst synthesizing research for an investment committee"
            },
            {
                "role": "user",
                "content": f"""Synthesize these raw findings into a comprehensive investment summary for {ticker}

                Raw findings from document analysis:
                {combined_points}

                Create a final summary with:
                1. business_description: Clear 1-2 sentence overview of core business
                2. material_points: Select the 1-5 MOST important facts for investment decisions
                3. investment_industry: List 1-3 industries the company is most exposed to
                4. investment_exposure: List 1-3 business or investment themes that are most relevant this company e: E-commerce transition, AI chip demand cycle
                - Eliminate redundant or overlapping points
                - Prioritize recent strategic, financial, and competitive information
                - Order by importance to stock valuation"""
                }
            ]
        )
        print("got to parsing")
        # Parse the final summary
        final_summary = json.loads(final_response['message']['content'])
        final_summary['status']= "success"
        final_summary['chunks_processed']= len(chunk_nodes)
        final_summary['total_key_points_found']= len(all_key_points)
        #print(final_summary)
        return final_summary

    except Exception as e:
        print(e)
        return {
            "status": "error",
            "message": f"Final synthesis failed: {e}",
            "ticker": ticker,
            "raw_points": all_key_points
        }


print(f"Analyzing document for WMT ({len(s.split())} words)...")
result = analyze_chunks(s,'WMT')
print(result)



```

```python
# Query for documents that have NOT been summarized yet
todo_cursor = collection. find({
    "wiki_content": {"$exists": True},
    "SUMMARY_material_points": {"$exists": False}
})


for row in todo_cursor: #if no summary yet
    ticker = row[ 'ticker']
    s = row[ 'wiki_content']
    print(f"\n\nAnalyzing document for {ticker} ({len(s.split())} words)...")
    result = analyze_chunks(s, ticker)

    SUMMARY_material_points = "\n".join(result.get('material_points',[]))
    SUMMARY_business_description = result.get('business_description','')
    SUMMARY_investment_industry = "\n".join(result.get('investment_industry',[]))
    SUMMARY_investment_exposure = "\n".join(result.get( 'investment_exposure',[]))

    collection. update_one(
        {"_id": row['_id']},
        {"$set": {
            "SUMMARY_material_points": SUMMARY_material_points,
            "SUMMARY_business_description": SUMMARY_business_description,
            "SUMMARY_investment_industry": SUMMARY_investment_industry,
            "SUMMARY_investment_exposure": SUMMARY_investment_exposure
        }})

    print(f"\n📊 COMPANY: {result.get('company_name', 'N/A')}")

    print(f"\n⚡ MATERIAL POINTS:")
    for i, point in enumerate(result.get('material_points', []), 1):
        print(f"   {i}. {point}")

    print(f"\n📝 BUSINESS DESCRIPTION:")
    print(f"  {result.get('business_description', 'N/A')}")

    print(f"\n🎯 INVESTMENT INDUSTRY:")
    for industry in result.get('investment_industry', []):
        print(f"   • {industry}")

    print(f"\n💼 INVESTMENT EXPOSURE:")
    for exposure in result.get('investment_exposure', []):
        print(f"   • {exposure}")

    print(f"Chunks Processed: {result.get('chunks_processed',0)}")
    print(f"Key Points Found: {result.get('total_key_points_found'),[]}")
```
